# 데이터 분석 
## 분석주제1 : 대출이 있는 사람이라면 은행 삼품에 잘 가입하지 않을 것이다.

```python
data.columns
```

```python
# 가입여부 칼럼 : y
grouped=data.groupby('y')
```

```python
yes_group=grouped.get_group('yes')
no_group=grouped.get_group('no')
```

```python
yes_group.head()
```

```python
no_group.head()
```

```python
yes=yes_group.loan.value_counts()
no=no_group.loan.value_counts()
```

```python
yes=yes/yes.sum()
yes.name='y_yes'
no=no/no.sum()
no.name='y_no'
```

```python
yes
```

```python
no
```

```python
pd.concat([yes,no], axis=1)
```

- 가입한 그룹의 대출 비중이 가입하지 않은 그룹보다 0.005 더 작다

### 본격 분석 시작

```python
pivot=pd.pivot_table(data, values='age', index='y', columns='job', aggfunc='mean')
pivot
```

```python
# 가입한 사람의 연령과 가입하지 않은 사람의 연령 차
pivot.loc['diff']=pivot.loc['yes']-pivot.loc['no']
pivot
```

```python
# diff 기준으로 내림차순 정렬
result=pivot.sort_values('diff', axis=1, ascending=False)
result
```

```python
result.loc['diff'].index
```

```python
plt.figure(figsize=(15,8))
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
sns.barplot(x=result.loc['diff'].index, y=result.loc['diff'].values)
```
