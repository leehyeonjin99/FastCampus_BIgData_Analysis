```python
from glob import glob
file_names=glob("cafe_data/*.csv")
total=pd.DataFrame()
for file_name in file_names:
    temp=pd.read_csv(file_name)
    total=pd.concat([total, temp])
total.reset_index(inplace=True, drop=True)
total
```

# 3. 커피전문점 별 비율 비교하기 (주요 브랜드 위주로)

2020년 10월 기준 커피전문점 평판 순위
(source : http://www.fntoday.co.kr/news/articleView.html?idxno=264802)
1. 스타벅스
2. 투썸플레이스
3. 메가커피

### 1) 전체 커피전문점 내 주요 커피 브랜드 입점 비율

```python
print("**** 전국 커피전문점중 주요 5대 커피브랜드 입점 비율 ****")
print("주요 5대 커피브랜드 전국 입점 비율 : %.3f%%" %((len(df_starbucks)+len(df_twosome)+len(df_megacoffee)+len(df_ediya)+len(df_paikdabang))/len(df_coffee)*100))
print("1. 스타벅스 : %.3f%%" % (len(df_starbucks)/len(df_coffee)*100) )
print("2. 투썸플레이스 : %.3f%%" % (len(df_twosome)/len(df_coffee)*100))
print("3. 메가커피 : %.3f%%" % (len(df_megacoffee)/len(df_coffee)*100))
print("4. 이디야 : %.3f%%" % (len(df_ediya)/len(df_coffee)*100))
print("5. 빽다방 : %.3f%%" % (len(df_paikdabang)/len(df_coffee)*100))
```

### 2) 서울 커피전문점 내 주요 커피브랜드 입점 비율

```python
print("스타벅스 : %.3f%%" % (len(df_seoul_starbucks)/len(df_seoul_coffee)*100))
print("이디야 : %.3f%%" % (len(df_seoul_ediya)/len(df_seoul_coffee)*100))
print("커피빈 : %.3f%%" % (len(df_seoul_coffeebin)/len(df_seoul_coffee)*100))
print("투썸플레이스 : %.3f%%" % (len(df_seoul_twosome)/len(df_seoul_coffee)*100))
print("빽다방 : %.3f%%" % (len(df_seoul_paikdabang)/len(df_seoul_coffee)*100))
print("할리스 : %.3f%%" % (len(df_seoul_hollys)/len(df_seoul_coffee)*100))
print("메가커피 : %.3f%%" % (len(df_seoul_megacoffee)/len(df_seoul_coffee)*100))
```

### 3) 각 커피브랜드별 서울 입점 비율

```python
print("**** 주요 5대 커피브랜드별 서울 입점 비율 ****")
print("1. 스타벅스 : %.3f%%" % (len(df_seoul_starbucks)/len(df_starbucks)*100))
print("2. 투썸플레이스 : %.3f%%" % (len(df_seoul_twosome)/len(df_twosome)*100))
print("3. 메가커피 : %.3f%%" % (len(df_seoul_megacoffee)/len(df_megacoffee)*100))
print("4. 이디야 : %.3f%%" % (len(df_seoul_ediya)/len(df_ediya)*100))
print("5. 빽다방 : %.3f%%" % (len(df_seoul_paikdabang)/len(df_paikdabang)*100))
```

```python
# 각 구별로 스타벅스가 얼마나 있는지 확인합니다.
starbucks_gu=df_seoul_starbucks.groupby('시군구명')['상호명'].count().to_frame().sort_values(by='상호명',ascending=False)
starbucks_gu=starbucks_gu.reset_index()
starbucks_gu=starbucks_gu.set_index('시군구명')
starbucks_gu
```

```python
# 주요 5대 커피브랜드 서울 입점 비율을 시각화합니다.
starbucks_rate=(len(df_seoul_starbucks)/len(df_starbucks)*100)
twosome_rate=(len(df_seoul_twosome)/len(df_twosome)*100)
megacoffee_rate=(len(df_seoul_megacoffee)/len(df_megacoffee)*100)
ediya_rate=(len(df_seoul_ediya)/len(df_ediya)*100)
paikdabang_rate=(len(df_seoul_paikdabang)/len(df_paikdabang)*100)

X=['스타벅스','투썸플레이스','메가커피','이디야','빽다방']
Y=[starbucks_rate, twosome_rate, megacoffee_rate, ediya_rate, paikdabang_rate]
plt.figure(figsize=(12,12))
plt.title("주요 5대 커피브랜드 서울 입점 비율", fontdict={'fontsize':20})
sns.barplot(x=X,y=Y)
plt.savefig("coffee_barplot.png")
plt.show
```
